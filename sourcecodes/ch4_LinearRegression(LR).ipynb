## Relationship between sleep hours, study hours, and score.

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
train_data = np.array([[8,3], [7,7], [4,13],[8,7],[10,5],[9,10]])
y_data = np.array([5,78,92,98,66,97])
```

```python
test_data = np.array([[10,2], [5,5], [0, 10]])
```

```python
# sleep hours vs. score
sleep_hours = np.array([train_data[x][0] for x in range(len(train_data))])
sleep_hours

```

```python
plt.scatter(sleep_hours, y_data)
plt.xlabel('sleep hours')
plt.ylabel('score')
plt.show()
```

```python
# study hours vs. score
study_hours = np.array([train_data[x][1] for x in range(len(train_data))])
study_hours
```

```python
plt.scatter(study_hours, y_data)
plt.xlabel('study hours')
plt.ylabel('score')
plt.show()
```

```python
# sleep hours vs. study hours vs. score
# 3d scatter

fig = plt.figure()
score_plt = fig.add_subplot(projection='3d')
score_plt.scatter(sleep_hours, study_hours, y_data)
score_plt.set_xlabel('sleep hours')
score_plt.set_ylabel('study hours')
score_plt.set_zlabel('score')
```

```python
model = LinearRegression()
model.fit(train_data, y_data)
```

```python
test_data
```

```python
model.predict(test_data)
```

```python
# get weights
model.coef_
```

```python
# get bias
model.intercept_
```

```python
fig2 = plt.figure()
ax = fig2.add_subplot(projection='3d')
ax.scatter(sleep_hours, study_hours, y_data)
ax.set_xlabel('sleep hours')
ax.set_ylabel('study hours')
ax.set_zlabel('sore')
w1 = np.array([model.coef_[0]])
w2 = np.array([model.coef_[1]])
bias = np.array([model.intercept_])
ax.plot_surface(w1,w2,np.array([w1*sleep_hours + w2*study_hours + bias]), color='black')
plt.show()
```

```python
# sleep hours vs. score
plt.scatter(sleep_hours, y_data)
plt.plot(sleep_hours, sleep_hours*w1+bias)
plt.show()
```

```python
# study hours vs. score
plt.scatter(study_hours, y_data)
plt.plot(study_hours, study_hours*w2 + bias)
```

```python
print("==== result of learning with 3 high-corr properties ====")
print("score of trainning data : ", model2.score(train_high_corr_property, train_diabetes2))
print("score of test data : ", model2.score(test_high_corr_property, test_diabetes2))
```

## life expentancy prediction using Linear Regression


## csv 파일 불러오기

```python
import pandas as pd

path = "/content/life_expectancy.csv"
life = pd.read_csv(path)
print(life.head(3))
```

```python
print('description of life expentancy dataset')
print(life.describe())
```

```python
print('columns of life data')
print(life.columns)
```

## get correlation and draw a heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(rc={'figure.figsize':(22,20)})
corr_matrix = life.corr().round(2)
sns.heatmap(corr_matrix, annot=True)
```

## Which have high-corr with life expectancy?

```python
print("coefficient to life expentancy")
print(life.corr().round(3)['Life expectancy'])
```

```python
corr = life.corr().round(3)['Life expectancy']
corr = corr.abs()
corr.sort_values()
```

## seaborn 쌍 그래프 시각화해보기

```python
lf = life[['Life expectancy', 'Schooling', 'Income composition of resources', 'Adult mortality', 'BMI', 'HIV/AIDS']]
sns.pairplot(lf)
plt.show()
```

## drop non data

```python
# check sum of non data

lf.isna().sum()
```

```python
life.dropna(inplace=True)
lf = life[['Life expectancy', 'Schooling', 'Income composition of resources', 'Adult mortality', 'BMI', 'HIV/AIDS']]
X = lf
y = life['Life expectancy']
```

## making a Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LinearRegression()
model.fit(X_train, y_train)
print('score of Linear Regression model : ', model.score(X_test, y_test).round(3))
```
